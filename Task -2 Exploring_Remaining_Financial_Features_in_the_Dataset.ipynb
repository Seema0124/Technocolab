```python
def plot_loghist(new, bins):
  hist, bins = np.histogram(new, bins=20)
  logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
  plt.hist(new, bins=logbins)
  plt.xscale('log')

plot_loghist(np.random.rand(200), 10)
```
